### True/Analytical PDF


## using `numpy`

```python
np.random.seed(42)
```

```python
n_sample = np.random.normal(loc=0, scale=1, size=10000)

plt.hist(n_sample, bins=20, density=True, color='blue', linewidth = 1.1)

x = np.linspace(-4, 4, 1000)
pdf = (1/np.sqrt(2*np.pi))*np.exp(-x**2/2)
plt.plot(x, pdf, color='red')

print(f"Sample mean: {np.mean(n_sample)}")
print(f"Sample variance: {np.var(n_sample)}")
plt.show()
```

## using `scipy`

```python
n_sample = np.random.normal(loc=0, scale=1, size=10000)

plt.hist(n_sample, bins=20, density=True, color='blue', linewidth = 1.1)

x = np.linspace(-4, 4, 1000)
pdf = stats.norm.pdf(x, 0, 1)
plt.plot(x, pdf, color='red')

print(f"Sample mean: {np.mean(n_sample)}")
print(f"Sample variance: {np.var(n_sample)}")
plt.show()


print(f"Sample mean: {stats.tmean(n_sample)}")
print(f"Sample variance: {stats.tvar(n_sample)}")
```

- Compute the true "analytical" PDF using [scipy.stats.norm.pdf](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html). Plot the true PDF on the top of the histogram figure.
- Compute the CDF using [scipy.stats.norm.cdf](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), then plot it.

```python
n_sample = np.random.normal(loc=0, scale=1, size=10000)

plt.hist(n_sample, bins=20, density=True, color='blue', linewidth = 1.1)

x = np.linspace(-4, 4, 1000)
pdf = stats.norm.pdf(x, 0, 1)
plt.plot(x, pdf, color='red')

cdf = stats.norm.cdf(x, 0, 1)
plt.plot(x, cdf, color='brown')

print(f"Sample mean: {np.mean(n_sample)}")
print(f"Sample variance: {np.var(n_sample)}")
plt.show()
```

C) Repeat (a) for an exponential random variable with $\lambda = 1$ using [numpy.random.exponential](https://numpy.org/doc/stable/reference/random/generated/numpy.random.exponential.html), and [scipy.stats.expon](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html#scipy.stats.expon)

```python
lambda_value = 1
n_sample = np.random.exponential(scale=1/lambda_value, size=1000)
plt.hist(n_sample, bins=30, density=True, color='blue', linewidth = 1.1)
x = np.linspace(0, max(n_sample), 1000)
pdf = stats.expon.pdf(x, scale=1/lambda_value)
plt.plot(x, pdf, color='red', linewidth=2, label='PDF')

print(f"Sample mean: {np.mean(n_sample)}")
print(f"Sample variance: {np.var(n_sample)}")

plt.legend()
plt.show()
```

```python
lambda_value = 1
n_sample = np.random.exponential(scale=1/lambda_value, size=1000)
plt.hist(n_sample, bins=30, density=True, color='blue', linewidth = 1.1, label='Histogram')
x = np.linspace(0, max(n_sample), 1000)
pdf = stats.expon.pdf(x, scale=1/lambda_value)
cdf = stats.expon.cdf(x, scale=1/lambda_value)
plt.plot(x, pdf, color='red', linewidth=2, label='PDF')
plt.plot(x, cdf, color='brown', linewidth=2, label='CDF')

plt.legend()
plt.show()
```

e) Repeat for an binomial random variable with $n = 5$ and $p = 0.5$

```python
n = 5
p = 0.5

n_sample = np.random.binomial(n, p, size=1000)

plt.hist(n_sample, bins=n, density=True, color='blue', linewidth = 1.1, label='Histogram')
x = np.arange(0, n+1)
pmf = stats.binom.pmf(x, n, p)
plt.vlines(x, 0, pmf, color='red', linewidth=2, label='PMF')

print(f"Sample mean: {np.mean(n_sample)}")
print(f"Sample variance: {np.var(n_sample)}")

plt.legend()
plt.show()
```

```python
n = 5
p = 0.5

n_sample = np.random.binomial(n, p, size=1000)

# plt.hist(n_sample, bins=5, density=True, color='blue', linewidth = 1.1, label='Histogram')
x = np.arange(0, n+1)
pmf = stats.binom.pmf(x, n, p)
cdf = stats.binom.cdf(x, n, p)
plt.vlines(x, 0, pmf, color='red', linewidth=2, label='PMF')
plt.step(x,cdf,where = 'post',color = 'b', label='CDF')

print(f"Sample mean: {np.mean(n_sample)}")
print(f"Sample variance: {np.var(n_sample)}")

plt.xticks(x)
plt.legend()
plt.show()
```

# Problem 3: Practicing the Central Limit Theorem
Use Python to simulate the Central Limit Theorem for the following random i.i.d variables and check that the higher the number of number the variables, the more the mean of subsets behave like a gaussian random variable.


```python
means_5 = []
for _ in range(1000):
    sample = np.random.uniform(0, 1, 5)
    means_5.append(np.mean(sample))

plt.hist(means_5, bins=30, density=True, alpha=0.6, color='b')

means_50 = []
for _ in range(1000):
    sample = np.random.uniform(0, 1, 50)
    means_50.append(np.mean(sample))


plt.hist(means_50, bins=100, density=True, alpha=0.6, color='b')
plt.show()
```
