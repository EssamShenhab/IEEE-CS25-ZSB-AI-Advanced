```python
import pandas as pd
```

```python
df = pd.read_csv("adult_data.csv")
df.head()
```

```python
df.info()
```

## How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.

```python
race_count = df["race"].value_counts()
race_count
```

## What is the average age of men?

```python
average_age_men = df["age"].mean()
average_age_men
```

## What is the percentage of people who have a Bachelor's degree?

```python
df.education.value_counts()
```

```python
# What is the percentage of people who have a Bachelor's degree?
percentage_bachelors = (df['education'] == 'Bachelors').mean() * 100
percentage_bachelors
```

## With `Bachelors`, `Masters`, or `Doctorate`

```python
higher_education = df[df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
higher_education
```

## Without `Bachelors`, `Masters`, or `Doctorate`

```python
lower_education = df[~df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
lower_education
```

## What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?

```python
higher_education_rich = round(sum(higher_education['salary'] == '>50K')/higher_education['salary'].count() * 100 , ndigits=2)
higher_education_rich
```

## What percentage of people without advanced education make more than 50K?

```python
lower_education_rich = round(sum(lower_education['salary'] == '>50K')/lower_education['salary'].count() * 100 , ndigits=2)
lower_education_rich
```

## What is the minimum number of hours a person works per week (hours-per-week feature)?

```python
min_work_hours = df["hours-per-week"].min()
min_work_hours
```

## What percentage of the people who work the minimum number of hours per week have a salary of >50K?

```python
num_min_workers = df[df["hours-per-week"] == min_work_hours]
num_min_workers
```

```python
rich_percentage = round(sum(num_min_workers['salary'] == '>50K') / num_min_workers['salary'].count() * 100, 2)
rich_percentage
```

## What country has the highest percentage of people that earn >50K?

```python
country_counts = (df[df["salary"] == '>50K']).groupby('native-country').size() / df.groupby('native-country').size() * 100
highest_earning_country = country_counts.idxmax()
highest_earning_country_percentage = round(country_counts.max(), 2)
highest_earning_country, highest_earning_country_percentage
```

## Identify the most popular occupation for those who earn >50K in India.


```python
india50K = df[(df["native-country"] == "India") & (df['salary'] == '>50K')]
top_IN_occupation = india50K['occupation'].value_counts().idxmax()
top_IN_occupation
```
